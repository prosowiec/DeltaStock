```python
import scraper
import pandas as pd
import numpy as np
```

```python
url = 'http://accounting.auditchain.finance/reporting-scheme/us-gaap/fac/Rules_Mapping/ConceptMap_General-mapping.html'
conn = scraper.APIconnector(url)
conn.get_request()
```

```python
# find_all("article", {"class":"ooa-1t80gpj ev7e6t818"})
soup = conn.make_soup()
table = np.array([x.text for x in soup.find_all('td', {"class" : "Row"})])
table = np.matrix(np.array_split(table, table.shape[0] / 5))
table
```

```python
headers = np.array([x.text for x in soup.find_all('th')[:5]])

headers
```

```python
dataframeDic = {}
for col in range(len(headers)):
    dataframeDic[headers[col]] = table[:,col].flatten().tolist()[0]
    
dataframeDic
```

```python
df = pd.DataFrame(dataframeDic)

df
```

```python
df.to_csv('SECmapping.csv', index = False)
```
